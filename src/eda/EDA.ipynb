```python
import wandb
import numpy as np
import pandas as pd
import ydata_profiling
```

```python
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = wandb.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)
print("DONE")
```

```python
profile = ydata_profiling.ProfileReport(df)
#profile.to_notebook_iframe()
df.info()
```

```python
# Drop outliers
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df.info()
```

```python
run.finish()
```

```python
# Confirm price column is within valid range
print("Min price:", df["price"].min())
print("Max price:", df["price"].max())

if df["price"].min() >= 0 and df["price"].max() <= 350:
    print("Prices are within the defined range.")
else:
    print("Prices are outside expected bounds!")

# 3. Confirm last_review is datetime type
print("\n=== last_review Type Check ===")
if np.issubdtype(df["last_review"].dtype, np.datetime64):
    print("Values of 'last_review' are of datetime type.")
else:
    print("Values for 'last_review' are NOT datetime. Current dtype:", df["last_review"].dtype)

# 4. (Optional) Regenerate profile report
regenerated_profile = ydata_profiling.ProfileReport(df)
regenerated_profile.to_file("profile_after_fixes.html")
print("\nGenerated: profile_after_fixes.html")
```
